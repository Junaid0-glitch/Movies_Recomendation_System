# Merging Both the datasets

```python
movies = movies.merge(creds, left_on="id", right_on="movie_id")
movies.head(3)
```

```python
movies.columns.tolist()
```

```python
movies = movies[['movie_id', 'title_x', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies = movies.rename(columns={'title_x': 'title'})
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.duplicated().sum()
```

# Data Cleaning & preprocessing

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(2)
```

```python
def convert6(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 6:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert6)
```

```python
movies.head(3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i["name"])
            break
    return L
```

```python
movies["crew"] = movies["crew"].apply(fetch_director)
```

```python
movies.head(3)
```

```python
movies["overview"] = movies["overview"].fillna("")
movies["overview"] = movies["overview"].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace (" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace (" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace (" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace (" ", "") for i in x])
```

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies["overview"]
movies.head()
```

```python

new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python

new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

# Feature extraction using SentenceTransformer

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
vector = model.encode(new['tags'].tolist(), normalize_embeddings=True)
```

```python
vector.shape
```

```python

from sklearn.metrics.pairwise import cosine_similarity
```

```python

similarity = cosine_similarity(vector)
```

```python

similarity[0]
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
